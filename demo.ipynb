```python
import sympy
import numpy as np
from sympy.abc import x
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as p
```

```python
polynomial = p.Polynomial([25, 4, 3, 0, 2, 1])
sympy.init_printing()
print(sympy.Poly(reversed(polynomial.coef), x).as_expr())
```

```python
positive_and_even_x = {x for x in range(-10, 20) if x % 2 == 0 and x >= 0}
print(positive_and_even_x)
print(len(positive_and_even_x))
print(2 in positive_and_even_x)
print(1 in positive_and_even_x)
```

```python
set1 = {0, 1, 2, 3, 4}
set2 = {5, 6, 7, 8, 9}

print(len(set1))
print(2 in set1)
print(9 in set1)
print({5, 6}.issubset(set2))
print(set1.union(set2))
print(set1.difference(set2))
print(set2.difference(set1))
print(set1.symmetric_difference(set2))
```

```python
def plot_function(f, min_x = -10, max_x = 10, n_values = 100):
    x = np.linspace(min_x, max_x, n_values)
    y = f(x)
    plt.plot(x, y)
    plt.show()
```

```python
plot_function(lambda x: np.sin(x), -10, 10, 1000)
plt.show()
plot_function(lambda x: np.cos(x), -10, 10, 1000)
plt.show()
plot_function(lambda x: np.tan(x), -10, 10, 1000)
plt.show()
```

```python
def graphing_unit_circle(f, min_x, max_x, n_values):
    plt.gca().set_aspect("equal")
    x = np.linspace(min_x, max_x, n_values)
    y = f(x)
    plt.plot(x, y)
```

```python
graphing_unit_circle(lambda x: np.sqrt(1 - x ** 2), -1, 1, 2000)
graphing_unit_circle(lambda x: -np.sqrt(1 - x ** 2), -1, 1, 2000)
plt.show()
```

```python
r = 1
phi = np.linspace(0, 2 * np.pi, 100)
x = r * np.cos(phi)
y = r * np.sin(phi)
plt.plot(x, y)
plt.gca().set_aspect("equal")
plt.show()
```
